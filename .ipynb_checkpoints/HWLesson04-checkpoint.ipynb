```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

17.6.5

```python
x = np.linspace(-15, 15, 200)
y = np.linspace(-15, 15, 200)
x, y = np.meshgrid(x, y)
a = -2.
b = 1.
c = 0.
d = -2.
e = -5.
fig = plt.figure(figsize=(10,6))
plt.axis([-15, 15, -15, 15])
plt.contour(x, y,(a*x**2 + b*y**2 + c*x + d*y + e), [1], colors='r')
plt.show()
```
