##### 2. Задание 
Докажите, что при ортогональном преобразовании сохраняется расстояние между точками.
 

квадрат расстояния между точками равен Rˆ2 = (x0-x1)ˆ2 + (y0-y1)ˆ2

возьмем x0-x1 за x, а y0-y1 за y, т.е. Rˆ2 = xˆ2 + yˆ2, соответсвенно большие X и Y

при линейном преобразовании X0-X1 и Y0-Y1 теряют коэффициенты a13 и а23, т.е.

X = a11*x + a12*y

Y = a21*x + a22*y

чтобы выяснить при каких условиях Rˆ2 от x, y = Rˆ2 от X, Y решим уравнение xˆ2 + yˆ2 = Xˆ2 + Yˆ2

подставляем вместо X и Y

(a11*x + a12*y)ˆ2 + (a21*x + a22*y)ˆ2 и раскроем скобки

a11ˆ2*xˆ2 + 2*a11*a12*xy + a12ˆ2*yˆ2 + a21ˆ2*xˆ2 + 2*a21*a22*xy +a22ˆ2*yˆ2 = xˆ2 + yˆ2, приравняем все 0 и упростим

xˆ2*(a11ˆ2 + a21ˆ2 - 1) + 2xy*(a11*a12 + a21*a22) + yˆ2*(a12ˆ2 + a22ˆ2 - 1) = 0, соответственно равенство всегда

будет верно при одновременном исполнении трех условий:

a11ˆ2 + a21ˆ2 = 1

a12ˆ2 + a22ˆ2 = 1

a11*a12 + a21*a22 = 0

что и является условием ортогонального преобразования.

##### 4. Задание (в программе)
Решите систему уравнений:

y = x2 – 1

exp(x) + x∙(1 – y) = 1

```python
x = np.linspace(-5, 10, 200)
y = np.linspace(-5, 20, 200)
x, y = np.meshgrid(x, y)
# plt.axis([-2, 6, -1.75, 3.75])
plt.contour(x, y, (x**2 - 1 - y), [1], colors='r')
plt.contour(x, y, (np.exp(x) + x * (1 - y) -1), [1], colors='b')
plt.show()
```

```python
from scipy.optimize import fsolve
```

```python
def equations(p):
    x, y = p
    return (x**2 - 1 - y, np.exp(x) + x * (1 - y) -1)

x1, y1 = fsolve(equations, (-2, 1))
x2, y2 = fsolve(equations, (0, 0))
x3, y3 = fsolve(equations, (2.5, 3))
x4, y4 = fsolve(equations, (5, 12))

print(x1, y1)
print(x2, y2)
print(x3, y3)
print(x4, y4)
```
